<b>2D convolution</b><br>
Kernel W shape: (hxw) <br>

$$y_{ij}=\sum_{k=0}^h \sum_{l=0}^w  X_{i+k,j+l}*w_{kl} $$ 


Output shape:<br>
Let, Input shape: X: (HxW) <br>
     Kernel shape: W: (hxw) <br>
     Output shape: y: (hhxww)<br>

output height would be, 
$$hh = \frac{H-h +2*pad}{stride}+1$$

Let, padding, pad=0 , filter size, hxw=(5x5), stride=1 and Input size: (28x28)<br>
So, $$hh = \frac{28-5+2*0}{1}+1=24$$
Similarly, 
$$ww = \frac{28-5+2*0}{1}+1=24$$

<b>Task 1: 2D convolution (hxwx1) </b> <br>
calculate output. <br>
Output should like below,<br><br>
[[ 100.   30.  -50.  -94.]<br>
 [ -50.  -50.  -10.  -85.]<br>
 [-230.   60.   30. -156.]<br>
 [-240.  140.  128. -189.]]

```python
X=np.array([[10,10,20,0,0, 4], [0, 0,50,50,20, 0], [10,60,0,0,30, 2], [120,20,40,40,0, 1], [70, 0,10,80,90, 0], [40, 40,0,0,8,10]])
K=np.array([[-1,0, 1],[-2,0,2], [-1,0,1]])
print(X)
print('kernel:')
print( K)
```

```python
#your code goes here.
def convolution(X,K,h,w,stride):
    y = []
    for i in range(0,h[0]-w[0]+1,stride):
        A = []
        for j in range(0,h[1]-w[1]+1,stride):
            s = X[i:i+w[0],j:j+w[1]]
            mul = np.multiply(s,K)
            z = np.sum(mul)
            A.append(z)
        y.append(A)
    return np.array(y)
```

```python
stride=1
y=convolution(X,K,X.shape,K.shape,stride)
print(y)
print("\nshape of y = ",y.shape)
```

<b>Task 2: ReLU</b><br>
$$y=max(0,x)$$

[[100.,  30.,   0.,   0.],<br>
[  0.,   0.,   0.,   0.],<br>
[  0.,  60.,  30.,   0.],<br>
[  0., 140., 128.,   0.]]

```python
#your code goes here.
def ReLu(y,shape):
    y_relu=[]
    for i in range(0,shape[0]):
        A=[]
        for j in range(0,shape[1]):
            A.append(max(0,y[i][j]))
        y_relu.append(A)
    return np.array(y_relu)
```

```python
y=ReLu(y,y.shape)
print(y)
print("\ny.shape ",y.shape)
```

<b>Task 3: MaxPooling</b><br>
Pool size=2, stride=2<br>
    
output should like below,<br>
[[100. -10.] <br>
 [140. 128.]] 

```python
def max_pool(y,pool,shape,stride):
    yy=[]
    for i in range(0,shape[0]-pool+1,stride):
        A=[]
        for j in range(0,shape[1]-pool+1,stride):
            s=y[i:i+pool,j:j+pool].flatten()
            mx=max(s)
            A.append(mx)
        yy.append(A)
    return np.array(yy)
```

```python
#your code goes here.
pool=2
stride=2
y2=max_pool(y,pool,y.shape,stride) 
print(y2)    
```

<b>Task 5: Convolution on Image. Multi channel</b>
1. Apply convolution
2. Plot the 8 images

```python
X=org.copy()
print('input image shape:',X.shape)
```

```python
n_C=8            #number of filters
n_C_prev=3       #number of channels in the input
f=5              #filter size
W=np.random.randn(n_C, f, f, n_C_prev)/9.0
print('filters shape:',W.shape)
```

```python
print('output shape should be: 216x216x8')
```

```python
#Task apply convolution, draw the 8 outputs seperately.
X=X.transpose()
print("X.shape ",X.shape)
print("W.shape ",W.shape)
```

```python
def convolution_multi_channel(X,K,h,w,stride):
    y=[]
    for i in range(0,h[1]-w[1]+1,stride):
        A=[]
        for j in range(0,h[2]-w[2]+1,stride):
            s=X[0:0+w[0],i:i+w[1],j:j+w[2]] 
            mul=np.multiply(s,K)
            z=np.sum(mul)
            A.append(z)
        y.append(A)
    return np.array(y)
```

```python
print(W[0].transpose().shape)
```

```python
img=[]
for i in range(0,n_C):
    t = W[i].transpose()
    h = X.shape
    w = t.shape
    stride = 1
    r = convolution_multi_channel(X,t,h,w,stride)
    img.append(r.transpose())
```

```python
fig = plt.figure(figsize=(20, 10))

for i in range(0,len(img)):
    im = img[i]
    fig.add_subplot(2,4, i+1)
    
    plt.title("convolution_features_"+str(i+1)) 
    plt.imshow(im,cmap="gray")
print('output shape = ',(img[0].shape,len(img)))
plt.show()
```
